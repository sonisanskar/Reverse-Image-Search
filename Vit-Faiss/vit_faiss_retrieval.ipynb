```python
from retrieval_system import ImageRetrievalSystem

retrieval = ImageRetrievalSystem(
    index_path="my_index.faiss",
    metadata_path="my_metadata.json"
)

query_path  = "/Users/sanskarsoni/Desktop/Reverse_Image_Search/Reverse_Image_Search/CNN_ANNOY/data/Footwear/Men/Images/images_with_product_ids/14490.jpg"
# Search for similar images
results = retrieval.search(query_path, k=5)

# Process results
all_paths=[]
for path, similarity in results:
    all_paths.append(path)
    print(f"Similar image: {path} (similarity: {similarity:.3f})")
    

image_paths = [query_path]+all_paths
# Iterate over each image path
num_images = len(image_paths)

# Create a subplot layout to display images side by side
plt.figure(figsize=(15, 5))  # Adjust the figsize to control the size of the display window

for i, image_path in enumerate(image_paths):
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    
    # Check if the image was read successfully
    if image is not None:
        # Convert the image from BGR to RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Add a subplot for each image
        plt.subplot(1, num_images, i + 1)  # 1 row, num_images columns, i+1 positioned subplot
        plt.imshow(image_rgb)
        plt.axis('off')  # Hide the axis labels
        plt.title(f"Image: {i+1}")  # Optional: set a title
    else:
        print(f"Error: Unable to read image at path {image_path}")

# Display all subplots
plt.tight_layout()
plt.show()
```
